### Multivariate Regression:

Independent Variables: *"House Size (sq.ft.)", "Number of Rooms", "Year of Construction"*

```python
X = data[['House Size (sq.ft.)', 'Number of Rooms', 'Year of Construction']]
Y = data['House Price']
```

```python
X1 = sm.add_constant(X)
reg = sm.OLS(Y, X1).fit()

reg.summary()
```
